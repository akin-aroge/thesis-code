```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
plt.rcParams['figure.figsize'] = 10, 10
```

```python
fig_dir = os.path.normpath(r"C:\Users\Fabusuyi\Documents\Work _UCT\PG\Write_Up\figs")
fig_save_dir = r"C:\Users\Fabusuyi\Documents\Work _UCT\PG\Write_Up\figs\setups"

def save_figure(fname, obj, dir=fig_save_dir, pad_inches=-0.5):
    """wrapper to save figure in desied directory"""
    path = os.path.join(os.path.normpath(fig_save_dir), fname)
    obj.savefig(path, pad_inches=pad_inches, bbox_inches='tight')
```

### fuel cell test station

```python
rel_path = os.path.normpath(r'random\active_load\IMG_20180530_194036.jpg')
fig_path = os.path.join(fig_dir, rel_path)

img = mpimg.imread(fig_path)
fig, ax = plt.subplots()
imgplot = ax.imshow(img)
t0 = ax.text(50, 500, 'active load\n device', fontsize=15)
t0.set_bbox(dict(facecolor='white', alpha=0.9, edgecolor='white'))

t1 = ax.text(600, 600, 'SCADA system', fontsize=15)
t1.set_bbox(dict(facecolor='white', alpha=0.9, edgecolor='white'))

t2 = ax.text(1600, 1000, 'host pc', fontsize=15)
t2.set_bbox(dict(facecolor='white', alpha=0.9, edgecolor='white'))

t3 = ax.text(1250, 2000, 'fuel cell\n assembly', fontsize=15)
t3.set_bbox(dict(facecolor='white', alpha=0.9, edgecolor='white'))

ax.axis('off');
save_figure('fuelcell_test_station.png', fig)
```
