```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib notebook

data_dir = r"C:\Users\Fabusuyi\Documents\Work _UCT\PG\Data\new_data\exp\my_data\data_fuelcon_random"  # change
```

```python
plt.rcParams['figure.figsize'] = 4,3
```

```python
def read_daq_data(data_dir, fname, columns=['time', 'value'], sep=' '):
    """this is a wrapper around the pd.read_csv to read my data format 
    from the fuel cell experiments
    
    params:
        data_dir: directory containing data
        fname: file_name
        columns: column names for the data
        sep: separator
    """
    
    
    data_path = os.path.join(os.path.normpath(data_dir), fname)
    
    return pd.read_csv(data_path, sep=sep, names=columns)
    
def plot_daq_data(df):
    fig, ax = plt.subplots()
    ax.plot(df['time'], df['value'])
```
