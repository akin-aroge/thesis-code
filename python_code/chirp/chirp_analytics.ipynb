```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

### The First Randle's

```python
def eis_dt_clean(data):
    
    """ cleans the data from the impedance for app"""
    col_names = ['z_re', 'z_im']
    data = data['101'].str.split('  ', 2, expand=True)
    data.drop(2, axis=1, inplace=True)
    data.columns = col_names
    
    return data
```

```python
col_names = ['z_re', 'z_im']

dt_cct0 = pd.read_csv('C:/Users/Aroogz/Desktop/Work _UCT/PG/Matlab Code/chirp/cct0.csv',
                       names=col_names)
dt_cct1 = pd.read_csv('C:/Users/Aroogz/Desktop/Work _UCT/PG/Matlab Code/chirp/cct1.csv',
                       names=col_names)
dt_cct13 = pd.read_csv('C:/Users/Aroogz/Desktop/Work _UCT/PG/Matlab Code/chirp/cct13.csv',
                       names=col_names )

dt_cct0_sim = pd.read_csv('C:/Users/Aroogz/Desktop/Work _UCT/PG/Python_Code/chirp/cct0.txt')
dt_cct1_sim = pd.read_csv('C:/Users/Aroogz/Desktop/Work _UCT/PG/Python_Code/chirp/cct1.txt')
dt_cct13_sim = pd.read_csv('C:/Users/Aroogz/Desktop/Work _UCT/PG/Python_Code/chirp/cct13.csv',
                          names=col_names,index_col=False )

dt_ccts = [dt_cct0, dt_cct1, dt_cct13]
dt_sims = [dt_cct0_sim, dt_cct1_sim]
dt_sims_clean = [0, 0, dt_cct13_sim]
for i,dt in enumerate(dt_sims):
    dt_clean = eis_dt_clean(dt)
    dt_sims_clean[i] = dt_clean
    
ax_label = ['circuit1', 'circuit2', 'circuit3']
fig, axs = plt.subplots(1,3, sharey=True)
fig.set_size_inches(10,3.5)
for ax, dt_cct, dt_cct_sim in zip(axs, dt_ccts, dt_sims_clean):
    
    x = dt_cct['z_re']
    y = dt_cct['z_im']
    x_sim = pd.to_numeric(dt_cct_sim['z_re'])
    y_sim = pd.to_numeric(dt_cct_sim['z_im'])
    
    ax.plot(x, y, '-*', label='chirp method')
    ax.plot(x_sim, y_sim, '-*', label='theoretical')
    
    ax.legend(prop={'size': 8})
    ax.grid()
    
axs[0].text(1,1, 'cct0')
axs[1].text(1,1, 'cct1')
axs[2].text(13,1, 'cct2')

fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top='off', 
                bottom='off', left='off', right='off')
plt.xlabel("Re(Z) $\Omega$")
plt.ylabel('$\minus$Im(Z) $\Omega$');
```
